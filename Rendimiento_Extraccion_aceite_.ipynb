# **Control de rendimiento de extraccion de aceite**

Este ejercicio es ideal para introducir float, int y la concatenación de strings.

**Planteamiento:**
Una planta procesadora de palma de aceite necesita calcular el Índice de Rendimiento de Extracción (IRE). El operario debe ingresar el peso total de los racimos de fruta fresca (RFF) procesados en kilogramos y la cantidad de aceite crudo obtenido en litros.

**Requerimientos:**

1. Solicitar el nombre del lote de producción (Tipo string).
2. Solicitar el peso de los racimos en kg (Tipo float). (Pr)
3. Solicitar los litros de aceite obtenidos (Tipo float). (Lt)
4. Calcular el rendimiento sabiendo que la densidad aproximada del aceite es de 0.92 kg/L. (r)

$$Pa=Lt * 0.92
$$
 $$r=(Pa/Pr) * 100$$

**Prueba del programa:**

Si se ingresan 1000 kg de racimos y se obtienen 200 litros de aceite, el programa debe informar que el rendimiento del lote es del 18.4%.

```python
# ENTRADA: Solicitar el nombre del lote de producción (Tipo string).
# Solicitar el peso de los racimos en kg (Tipo float). (Pr)
# Solicitar los litros de aceite obtenidos (Tipo float). (Lt)
nombre_lote = input("Ingrese el nombre del lote: ")
Pr = float(input("Ingrese el peso de los racimos en kg: "))
Lt = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))

# PROCESO: calcular peso del aceite (Pa) y el rendimiento (r)
Pa = Lt * 0.92
r = (Pa / Pr) * 100
#SALIDA: informar el rendimiento del lote
print("El rendimiento del lote es: ", r)
```

Aquí se explica cada línea del código para el cálculo del rendimiento de extracción de aceite:

*   `nombre_lote = input("Ingrese el nombre del lote: ")`
    *   Esta línea solicita al usuario que ingrese el nombre del lote de producción. La función `input()` lee la entrada del teclado y la almacena como una cadena de texto en la variable `nombre_lote`.

*   `Pr = float(input("Ingrese el peso de los racimos en kg: "))`
    *   Esta línea solicita al usuario el peso de los racimos de fruta fresca en kilogramos. La función `input()` obtiene la entrada como una cadena, y `float()` la convierte a un número de punto flotante (decimal) antes de asignarla a la variable `Pr` (Peso de racimos).

*   `Lt = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))`
    *   Similar a la línea anterior, esta solicita al usuario la cantidad de aceite crudo obtenido en litros. La entrada se convierte a un número de punto flotante y se guarda en la variable `Lt` (Litros de aceite).

*   `Pa = Lt * 0.92`
    *   Esta línea calcula el peso del aceite (Pa). Multiplica la cantidad de aceite en litros (`Lt`) por su densidad aproximada (0.92 kg/L). El resultado es el peso del aceite en kilogramos.

*   `r = (Pa / Pr) * 100`
    *   Esta línea calcula el rendimiento de extracción (`r`). Divide el peso del aceite obtenido (`Pa`) entre el peso de los racimos procesados (`Pr`) y luego multiplica el resultado por 100 para expresarlo como un porcentaje.

*   `print("El rendimiento del lote es: ", r)`
    *   Finalmente, esta línea imprime el resultado en la consola. Muestra el mensaje "El rendimiento del lote es: " seguido del valor calculado del rendimiento (`r`).
