### 1. Cálculo de la Humedad de un Producto

Este script calcula el porcentaje de humedad en base húmeda, algo vital para el secado de granos.

```
# Datos de entrada: peso inicial y peso seco
peso_inicial = 500.0  # gramos
peso_seco = 420.0     # gramos

# Fórmula de humedad base húmeda
humedad = ((peso_inicial - peso_seco) / peso_inicial) * 100

print(f"El porcentaje de humedad de la muestra es: {humedad}%")
```

### 2. Clasificación Automática de Calidad (Madurez)

Imagina que un sensor te da un valor de color o firmeza. Este código decide si el producto es apto para exportación.

```
# Escala de firmeza de 1 a 10
firmeza_fruta = 8.5

if firmeza_fruta >= 8:
    print("Estado: Óptimo para Exportación")
elif firmeza_fruta >= 5:
    print("Estado: Consumo Local")
else:
    print("Estado: Procesamiento (Mermeladas/Jugos)")
```

### 3. Estimación de Rendimiento por Hectárea

Ideal para planeación logística y proyecciones de ventas.

```
hectareas = 15
promedio_por_hectarea = 12.5  # toneladas

total_estimado = hectareas * promedio_por_hectarea

print(f"Producción total estimada: {total_estimado} Toneladas")
```
